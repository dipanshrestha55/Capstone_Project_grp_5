```python
# Import Dependencies
from tensorflow.keras.preprocessing import image
import tensorflow as tf
from tensorflow.keras import models, layers
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# Constants
BATCH_SIZE = 32
IMAGE_SIZE = 224
CHANNELS = 3
EPOCHS = 15
DATA_DIR = "mango_dataset"   # folder containing dataset
```

```python
# Load Dataset
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    DATA_DIR,
    seed=123,
    shuffle=True,
    image_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE,
)

class_names = dataset.class_names
print("Classes:", class_names)
num_classes = len(class_names)
print("Number of classes:", num_classes)
```

```python
# Split Dataset
def get_dataset_partitions_tf(ds, train_split=0.8, val_split=0.1, test_split=0.1, shuffle=True, shuffle_size=10000):
    assert (train_split + val_split + test_split) == 1
    ds_size = len(ds)
    if shuffle:
        ds = ds.shuffle(shuffle_size, seed=12)
    train_size = int(train_split * ds_size)
    val_size = int(val_split * ds_size)
    train_ds = ds.take(train_size)
    val_ds = ds.skip(train_size).take(val_size)
    test_ds = ds.skip(train_size).skip(val_size)
    return train_ds, val_ds, test_ds

train_ds, val_ds, test_ds = get_dataset_partitions_tf(dataset)
```

```python
# Cache, shuffle, prefetch
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds = val_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds = test_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
# Data Augmentation
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.2),
])

train_ds = train_ds.map(
    lambda x, y: (data_augmentation(x, training=True), y)
).prefetch(buffer_size=tf.data.AUTOTUNE)
```

```python
# Function to build a model using a pretrained base
def build_model(base_model_name):
    IMG_SHAPE = (IMAGE_SIZE, IMAGE_SIZE, CHANNELS)
    
    if base_model_name == "MobileNetV2":
        base_model = tf.keras.applications.MobileNetV2(
            input_shape=IMG_SHAPE, include_top=False, weights='imagenet')
    elif base_model_name == "ResNet101":
        base_model = tf.keras.applications.ResNet50(
            input_shape=IMG_SHAPE, include_top=False, weights='imagenet')
    elif base_model_name == "VGG16":
        base_model = tf.keras.applications.VGG16(
            input_shape=IMG_SHAPE, include_top=False, weights='imagenet')
    elif base_model_name == "DenseNet201":
        base_model = tf.keras.applications.DenseNet201(
            input_shape=IMG_SHAPE, include_top=False, weights='imagenet')
    else:
        raise ValueError("Unknown base model")

    base_model.trainable = False  # freeze pretrained weights

    inputs = tf.keras.Input(shape=IMG_SHAPE)
    x = layers.Rescaling(1./255)(inputs)
    x = base_model(x, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(64, activation='relu')(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = models.Model(inputs, outputs)
    model.compile(
        optimizer='adam',
        loss=tf.keras.losses.SparseCategoricalCrossentropy(),
        metrics=['accuracy']
    )
    return model
```

```python
# Compare models
model_names = ["ResNet101", "DenseNet201", "MobileNetV2", "VGG16",]
results = {}

for name in model_names:
    print(f"\nTraining {name}...")
    model = build_model(name)
    history = model.fit(train_ds, validation_data=val_ds, epochs=EPOCHS, verbose=1)
    
    test_loss, test_acc = model.evaluate(test_ds, verbose=0)
    print(f"{name} Test Accuracy: {test_acc*100:.2f}%")
    results[name] = test_acc*100
```

```python
# Show comparison
print("\n=== Model Comparison ===")
for model_name, acc in results.items():
    print(f"{model_name}: {acc:.2f}% accuracy")
```

```python
# plot bar chart
plt.figure(figsize=(6,4))
plt.bar(results.keys(), results.values(), color=['blue','green','orange','purple'])
plt.ylabel("Test Accuracy (%)")
plt.title("Pretrained Model Comparison on Mango Leaf Dataset")
plt.show()
```

```python
# Load DenseNet201 base
IMG_SHAPE = (IMAGE_SIZE, IMAGE_SIZE, CHANNELS)
base_model = tf.keras.applications.DenseNet201(
    input_shape=IMG_SHAPE,
    include_top=False,
    weights='imagenet'
)
```

```python
# Check if loaded successfully
print("DenseNet201 base model loaded successfully!")
print("Model summary:")
base_model.summary()
```

```python
# Optional: print first 5 layer names
print("\nFirst 5 layers of the base model:")
for i, layer in enumerate(base_model.layers[:5]):
    print(i, layer.name)
```

```python
# Freeze base model
base_model.trainable = False
```

```python
# Build the full model
inputs = tf.keras.Input(shape=IMG_SHAPE)
x = tf.keras.layers.Rescaling(1./255)(inputs)
x = base_model(x, training=False)
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(64, activation='relu')(x)
outputs = tf.keras.layers.Dense(len(class_names), activation='softmax')(x)

model = tf.keras.Model(inputs, outputs)
model.summary()
```

```python
# Compile the model
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=['accuracy']
)
```

```python
# Train the model
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10
)
```

```python
# Evaluate on test dataset
scores = model.evaluate(test_ds)
print("Test Loss:", scores[0])
print("Test Accuracy:", scores[1])
```

```python
# Plot Accuracy and Loss
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(range(10), acc, label='Training Accuracy')
plt.plot(range(10), val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(range(10), loss, label='Training Loss')
plt.plot(range(10), val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
# Prediction function
def predict(model, img):
    img_array = tf.expand_dims(img, 0)
    predictions = model.predict(img_array)
    predicted_class = class_names[np.argmax(predictions[0])]
    confidence = round(100 * np.max(predictions[0]), 2)
    return predicted_class, confidence

# Test predictions on few images
plt.figure(figsize=(15, 15))
for images, labels in test_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        predicted_class, confidence = predict(model, images[i].numpy())
        actual_class = class_names[labels[i]]
        plt.title(f"Actual: {actual_class}\nPredicted: {predicted_class}\nConfidence: {confidence}%")
        plt.axis("off")
```

```python
# Set a confidence threshold for unknown predictions
CONFIDENCE_THRESHOLD = 0.4  # 40%

for images_batch, labels_batch in test_ds.take(1):
    
    first_image = images_batch[0].numpy().astype('uint8')
    first_label = labels_batch[0].numpy()
    
    print("First image to predict")
    plt.imshow(first_image)
    plt.axis('off')
    plt.show()
    print("Actual label:", class_names[first_label])
    
    batch_prediction = model.predict(images_batch)
    predicted_probs = batch_prediction[0]
    predicted_index = np.argmax(predicted_probs)
    confidence = predicted_probs[predicted_index]
    
    if confidence < CONFIDENCE_THRESHOLD:
        predicted_label = "Unknown / Not a recognized leaf disease"
    else:
        predicted_label = class_names[predicted_index]
    
    print(f"Predicted label: {predicted_label}, Confidence: {confidence*100:.2f}%")
```

```python
# Save the model
models_dir = "../models"
os.makedirs(models_dir, exist_ok=True)

existing_versions = []
for name in os.listdir(models_dir):
    num_part = name.split('.')[0]
    if num_part.isdigit():
        existing_versions.append(int(num_part))

model_version = max(existing_versions + [0]) + 1
model.save(os.path.join(models_dir, f"{model_version}.keras"))

# Also save with specific name
model.save("mangos_densenet201.keras")

print(f"Model saved successfully as version {model_version} and as 'mangos_densenet201.keras'")
```

```python
from tensorflow.keras.utils import load_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt

# Function to predict a single image from file
def predict_image(model, img_path, img_size=(IMAGE_SIZE, IMAGE_SIZE), threshold=0.1):
    # Load image
    img = load_img(img_path, target_size=img_size)  # resize to model input
    img_array = img_to_array(img)
    
    # Show the image
    plt.imshow(img_array.astype('uint8'))
    plt.axis('off')
    plt.show()
    
    # Predict
    img_array_expanded = np.expand_dims(img_array / 255.0, axis=0)
    predictions = model.predict(img_array_expanded)
    predicted_index = np.argmax(predictions[0])
    confidence = predictions[0][predicted_index]
    
    if confidence < threshold:
        predicted_label = "Unknown / Not a recognized leaf disease"
    else:
        predicted_label = class_names[predicted_index]
    
    print(f"Predicted label: {predicted_label}, Confidence: {confidence*100:.2f}%")
    return predicted_label, confidence

# Example usage
img_path = "C:\\code\\MANGO PROJECT LATEST\\training\\mango_dataset\\Healthy\\20211231_161057 (Custom).jpg"
predict_image(model, img_path)
```
